```python
import pandas as pd
```

```python
dataset=pd.read_csv('heart.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
y=dataset['output']
```

```python
y
```

```python
x=dataset[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']]
```

```python
x
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape
```

```python
x_train
```

```python
y_train.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.optimizers import Adam
```

```python
model=Sequential()
```

```python
model.add(Dense(input_dim=13,units=12,activation='relu'))
```

```python
model.summary()
```

```python
model.add(Dense(units=8,activation='relu'))
```

```python
model.summary()
```

```python

model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=200)
```

```python
model.save('heart_pred1.h5')
```

```python
model.predict([[45,1,3,110,264,0,1,132,0,1.2,1,0,3],[63,1,3,145,233,1,0,150,0,2.3,0,0,1]])
```

```python
model.get_weights()
```

```python
dataset
```

```python
dataset.iloc[0:3]
```

```python
from tensorflow import keras
model = keras.models.load_model('heart_pred1.h5')
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.evaluate(x_train,y_train)
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,y_pred.round()))
```

```python
print(metrics.confusion_matrix(y_test,y_pred.round()))
```
